```python
#CLT is the foundation of inferential statistics — it's why we can make 
#predictions and run hypothesis tests on real-world data, even when that 
#data isn't normally distributed.
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("ready")
```

```python
# population generator
def create_population():
    initial_X_values = np.random.default_rng()

    rand_points_x = initial_X_values.integers(1,10,size=(1000))
    rand_points_x.tolist()
    return rand_points_x


population = create_population()
```

```python
# extract 30 random average values from the population 20 times
def calculate_CLT():
    means=[]

    for i in range(10000):
       means.append(np.mean(np.random.choice(population, size = 30)))

     
    counts, bin_edges, _ = plt.hist(means, bins=10,alpha = 0.7)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    plt.plot(bin_centers, counts, color='red', marker='o')
    plt.title("CLT Theorem")
    plt.show()
    
```

```python
calculate_CLT()
```
